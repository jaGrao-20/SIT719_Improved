```python
#import libraries

import pandas as pd
import os
from sklearn.metrics import confusion_matrix
#sklearn.metrics: Score functions, performance metrics, pairwise metrics and distance computations.
#confusion_matrix: Compute confusion matrix to evaluate the accuracy of a classification.
```

```python
link_to_data = 'https://raw.githubusercontent.com/SIT719/2020-S2/master/data/payment_fraud.csv'
dfp = pd.read_csv(link_to_data, on_bad_lines='skip')
```

```python
dfp
```

```python
df = pd.get_dummies(dfp, columns=['paymentMethod'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
#sklearn.model_selection: Tools for model selection, such as cross validation and hyper-parameter tuning.
#train_test_split: Split arrays or matrices into random train and test subsets.
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('label', axis=1), df['label'],
    test_size=0.33, random_state=17)
```

```python
X_train
y_train
```

```python
from sklearn.linear_model import LogisticRegression
#sklearn.linear_model: A variety of linear models
#LogisticRegression: This model implements regularised logistic regression using a set of available solvers
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
#sklearn.metrics: Score functions, performance metrics, pairwise metrics and distance computations.
#accuracy_score: In multilabel classification, this function computes subset accuracy: the set of labels predicted for a sample must exactly match the corresponding set of labels in y_true.
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
print(confusion_matrix(y_test, y_pred))
```
